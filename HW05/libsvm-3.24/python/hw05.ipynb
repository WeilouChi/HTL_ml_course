```python
import numpy as np
import pandas as pd
import scipy
from svmutil import *
```

```python
def Q15():
    y, x = svm_read_problem('train.txt',return_scipy=True)
    y = np.where(y!=3,0,y)
    m = svm_train(y,x,'-t 0 -c 10')
    p_label, p_acc, p_val = svm_predict(y, x, m)
    SVC =pd.DataFrame(m.get_sv_coef())
    SV = pd.DataFrame(m.get_SV()).fillna(0)
    tem = SVC.values*SV.values
    w = sum(tem)
    w =np.sqrt(np.sum(np.square(w)))
    return w
print(Q15())
```

```python
def Q20(): 
    from sklearn.utils import shuffle
    res = [0]*5
    for _ in range(1000):
        train_y, train_x = svm_read_problem('train.txt',return_scipy=True)
        train_x, train_y = shuffle(train_x, train_y)
        validation_x = train_x[:200]
        validation_y = train_y[:200]
        train_x = train_x[200:]
        train_y = train_y[200:]
        gam = [0.1,1,10,100,1000]
        maxACC = 0
        for i in range(len(gam)):
            m = svm_train(train_y,train_x,('-s 0 -t 2 -c 0.1 -g %s')%gam[i])
            p_label, p_acc, p_val = svm_predict(validation_y, validation_x, m)
            ACC, MSE, SCC = evaluations(validation_y, p_label)
            if ACC > maxACC:
                maxACC = ACC
                tem = i
        res[tem] += 1
        print("----------",_,"-----------")
    count = max(res)
    ans = res.index(count)
    return ans
Q20()
```
